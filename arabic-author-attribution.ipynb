```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

pd.set_option('max_colwidth', 400)
```

```python
import os
rootdir = 'D:\\DeepLearning\\Author-Attribution\\arabic-dataset'
rows = []
for subdir, dirs, files in os.walk(rootdir):
   
   count = 0
   for file in files:
      row = {}
      filename = os.fsdecode(file)
      
      if count ==10:
         count=0

      if filename.endswith('.properties'):
         with open(os.path.join(subdir, file),'r',encoding='utf-8') as f:
            data1 = f.readlines()
            row['author'] = data1[1].split('=')[1][0]
            with open(os.path.join(subdir, 'sample' + str(count)+'.txt'), encoding='utf-8') as f1:
               data = f1.read()
               row['text'] = str(data)
               count+=1
         rows.append(row)
```

```python
rows[0]
```

```python
df = pd.DataFrame.from_dict(rows, orient='columns')

df.shape
```

```python
df.head(40)
```

```python
import re

df['text']= df['text'].apply(lambda x: re.sub('[^ا-ي\s]','',x))
```

```python
df.head()
```

```python
MAX_FEATURES = 1000
NUM_CLASSES = 9
```

```python
from nltk.corpus import stopwords

X = df['text']
Y = df['author']

vectorizer = TfidfVectorizer(max_features=MAX_FEATURES,stop_words=stopwords.words('arabic'))
X = vectorizer.fit_transform(X).toarray()
Y = Y.to_numpy()
```

```python
X = np.array(X,dtype=np.float32)
Y = np.array(Y,dtype=np.int64)
X.shape , Y.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() 
Y = le.fit_transform(Y)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
features = vectorizer.get_feature_names_out()
df_post_processing = pd.DataFrame(X,columns=features).head()
df_post_processing.head()
```

```python
import numpy as np  

train_ratio = 0.7
validation_ratio = 0.15
test_ratio = 0.15

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=1 - train_ratio)

x_val, x_test, y_val, y_test = train_test_split(x_test, y_test, test_size=test_ratio/(test_ratio + validation_ratio)) 


print(x_train.shape, x_val.shape, x_test.shape)
print(y_train.shape, y_val.shape, y_test.shape)
```

# Mutinomial NB model

```python
nb = MultinomialNB()
nb.fit(x_train,y_train)
print(f"Accuracy of the model: {accuracy_score(y_test,nb.predict(x_test))}")
```

## HyperTuning to MultinomialNB()

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
'alpha': (1, 0.1, 0.01, 0.001, 0.0001, 0.00001),
'fit_prior':[True, False]
}

nb = MultinomialNB()
clf = GridSearchCV(nb, param_grid)
clf.fit(x_train,y_train)

print("Best Parameters: ", clf.best_params_)
print("Best Accuracy: {:.2f}".format(clf.best_score_))
```

```python
nb = MultinomialNB(alpha=1 , fit_prior=True)
nb.fit(x_train,y_train)
print(f"Accuracy of the model: {accuracy_score(y_test,nb.predict(x_test))}")
```

# Logistic Regression

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
print(f"Accuracy of the model: {accuracy_score(y_test,lr.predict(x_test))}")
```

# HyperTuning to LogisticRegression()

```python
lr = LogisticRegression(solver='liblinear')
param_grid = {
    'C': [0.001, 0.01, 0.1],
    'penalty': ['l1', 'l2']
}

# Create the GridSearchCV object
grid_search = GridSearchCV(lr, param_grid, scoring='roc_auc', cv=5)

# Fit the model to the training data
grid_search.fit(x_train, y_train)

# Print the best hyperparameters
print("Best hyperparameters:", grid_search.best_params_)
```

```python
lr = LogisticRegression(C=0.001,solver='liblinear')
lr.fit(x_train,y_train)
print(f"Accuracy of the model: {accuracy_score(y_test,lr.predict(x_test))}")
```
